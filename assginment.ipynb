```python
data = open("Desktop\customerdata.txt",'r')
data = data.readlines()
ctmdata = []
for line in data:
    temp = []
    for eles in line.split():
        temp.append(eles[:-1])
    ctmdata.append(temp)    
ctmdata    
```

```python
ctmdata = ctmdata[1:-1]
ctmdata
```

```python
# 1.How many orders did the site receive?

print(len(ctmdata))
```

```python
# 2.What was the total amount of the orders?
sum = 0
for i in ctmdata:
    sum += int(i[4])
print(sum)    
        
```

```python
# 3.List the names of the customers who ordered once and did not order again.
dir = {}
for i in ctmdata:
    if i[2]+' '+i[3] not in dir:
        dir[i[2]+' '+i[3]] = 0
    else:
        dir[i[2]+' '+i[3]] = int(dir[i[2]+' '+i[3]]) +1
users = []
for j in dir:
    if dir[j]==0:
        users.append(j)
print(users)
```

```python
# 4.Get a distribution of customers who ordered exactly once, exactly twice, and so on up to 4 orders and group the rest as 5 orders and abo
cnt_stats = [0]*5
for j in dir:
    if dir[j]>=5:
        cnt_stats[4] += 1
    elif dir[j]==4:
        cnt_stats[3] += 1
    elif dir[j]==3:
        cnt_stats[2] += 1
    elif dir[j]==2:
        cnt_stats[1] += 1
    elif dir[j]==1:
        cnt_stats[0] += 1
fin = [['No. of orders','Count of customers']]
for i in range(0,5):
    if i==4:
        fin.append([str(str(i+1)+'+'),cnt_stats[i]])
    else:
        fin.append([i+1,cnt_stats[i]])
fin
```
